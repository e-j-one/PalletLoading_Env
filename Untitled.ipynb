```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
result_dir = "./results/board/"
model_name = "FCDQN_1108_1629"

result_file = result_dir + model_name + ".npy"
result_ = np.load(result_file, allow_pickle=True)

rewards_ = result_[0]
losses_ = result_[1]

eplens_ = result_[2]
cumlens_ = np.cumsum(eplens_)

n_data = len(rewards_)


def smoothing(list_, n_smooth=500):
    smoothed_list_ = []
    for i in range(n_data):
        if i < n_smooth:
            smoothed_list_.append(np.mean(list_[:i+1]))
        else:
            smoothed_list_.append(np.mean(list_[i-n_smooth:i+1]))
    return smoothed_list_            


def plot_graph(x_values, y_values, color_, title_):
    plt.figure(figsize=(6,4))
    plt.plot(x_values, y_values, c=color_)
    plt.title(title_)
    plt.grid(True)
    
    
smoothed_rewards_ = smoothing(eplens_)
plot_graph(cumlens_, smoothed_rewards_, "tomato", "DQN Performance")
smoothed_losses_ = smoothing(losses_)
plot_graph(cumlens_, smoothed_losses_, "orange", "DQN Loss")
plt.show()
```
